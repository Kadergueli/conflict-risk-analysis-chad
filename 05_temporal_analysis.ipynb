```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/ACLED_CHAD_RESEARCH/data/cleaned_acled_chad.csv')
```

```python
# conversion de la date
df['EVENT_DATE'] = pd.to_datetime(df['EVENT_DATE'])
df.head()
```

```python
#creation des colonnes de temp
df['year'] = df['EVENT_DATE'].dt.year
df['month'] = df['EVENT_DATE'].dt.month
df['month_name'] = df['EVENT_DATE'].dt.month_name()
```

```python
#nombre d'événements par annee
year_counts = df.groupby('YEAR').size()

year_counts.plot(kind='line', marker='o', figsize=(10,5))
plt.title("Conflict Events by Year in Chad")
plt.xlabel("Year")
plt.ylabel("Number of Events")
plt.show()
```

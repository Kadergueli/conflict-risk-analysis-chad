```python

import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/ACLED_CHAD_RESEARCH/data/acled_chad.csv')
#vue des données existants
df.shape
df.columns
print(df.head())
```

```python
#elimination des cellules non indispensables
df.info()
df = df.drop(columns=['GWNO','EVENT_ID_CNTY','EVENT_ID_NO_CNTY','SOURCE','NOTES','ALLY_ACTOR_1','ACTOR1','ACTOR2','INTER1','INTER2','ALLY_ACTOR_2'])
```

```python
# nouvelle forme du dataframe
df.shape
df.head()
```

```python
#calcul statistique pour verifier les valeurs marquantes
df.describe()
df.isnull().sum()
```

```python
# elimination des doublons
df.drop_duplicates()
```

```python
# nouvelle dimension du dataframe
df.shape
df.head()
```

```python
df['FATALITIES'] = df['FATALITIES'].fillna(0)
```

```python
#regroupement par regions
df.groupby('ADMIN1').count()
```

```python
#verification final
df.info()
```

```python
#save
df.to_csv('/content/drive/MyDrive/ACLED_CHAD_RESEARCH/data/cleaned_acled_chad.csv', index=False)
```
